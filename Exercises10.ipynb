## Comprehending lists

As we have seen before, a list can be constructed with a `for` loop, or equivalently, with a list comprehension.

For example, suppose we want to calculate the squares of numbers from 1 to 20, excluding those that are divisible by 3 or 5.

With a for-loop:

````python
L = []
for x in range(1,21):
    if not (x%3==0 or x%5==0):
        L.append(x*x)
L
````

With a list comprehension:

````python
L = [x*x for x in range(1,21) if not (x%3==0 or x%5==0)]
L
````
