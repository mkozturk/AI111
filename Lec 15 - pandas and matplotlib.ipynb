# Plotting with pandas

*pandas* Series and Dataframes are based on NumPy arrays. They can work seamlessly with Numpy and Matplotlib functions.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
s = pd.Series([0.1, 0.3, 0.5, 1.1, 0.9, 0.85, 1.1], name="my data")
```

```python
plt.plot(s)
```

```python
temperature = [0.46622,2.58163,5.52012,9.27031,15.381,17.8839,23.015,23.067,21.515,14.4935,8.454,0.97317]
rainfall = [83.8037,76.1691,90.438,52.9172,59.2425,64.8199,5.82473,21.858,22.9251,77.3776,31.0791,29.1979]
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
climate = pd.DataFrame({"temperature":temperature,"rainfall":rainfall},index=months)
climate
```

We can extract individual columns and use them with matplotlib as usual:

```python
plt.plot(climate["temperature"])
plt.plot(climate["rainfall"])
```

pandas Dataframes provide a `plot()` method, which is very configurable and handles many details under the hood.

```python
climate.plot()
```

We can add plot elements using the familiar matplotlib syntax:

```python
climate.plot()
plt.title("Averages over years")
plt.xlabel("Month of the year")
plt.ylabel("Amount")
plt.grid(True)
```

Many of these settings can be passed to the `DataFrame.plot()` function directly:

```python
climate.plot(title="Averages over years", xlabel="Month of the year",ylabel="Amount", grid=True);
```

We can choose among many different kinds of plots with the `kind=` parameter.

A bar chart:

```python
climate.plot(kind="bar");
```

A scatter plot:

```python
climate.plot(kind="scatter", x="temperature", y="rainfall", grid=True)
```

We can generate a separate subplot for each column with `subplots=True`. We can also set the figure size:

```python
climate.plot(subplots=True,figsize=(6,8));
```

The layout of subplots can be set with the `layout=` parameter:

```python
climate.plot(subplots=True, kind="bar",layout=(1,2),figsize=(12,3));
```

Histograms of data, and kernel-density estimation plots can be generated in the same way:

```python
x = pd.Series(np.random.randn(1000))

x.plot(kind="hist",
       bins=20,     # divide the range of data into 20 intervals
       density=True, # plot percentages instead of counts in each bin
       figsize=(4,3)
      )
x.plot(kind="kde"); # a smoother version of a histogram
```

A *boxplot* is a special way to summarize data.

![](img/box-plot-construction.png)

On a Series or DataFrame, it can be generated over each column with the `kind="box"` parameter.

```python
grades = pd.read_csv("data_ex8_1.csv")
grades.plot(kind="box", title="Distribution of grades", ylabel="score");
```
