```python
import numpy as np
import matplotlib.pyplot as plt
```

# Subplots 

Plot the curve of $sin(t)$ vs $t$, and $cos(t)$ vs $t$ in vertical arrangement.

```python
t = np.linspace(0, 4*np.pi, 100)

plt.subplot(2,1,1) # 2 rows, 1 column, 1st plot
plt.plot(t, np.sin(t))

plt.subplot(2,1,2) # 2 rows, 1 column, 2nd plot
plt.plot(t, np.cos(t))
```

```python
t = np.linspace(0, 4*np.pi, 100)

plt.subplot(1,2,1) # 1 row, 2 columns, 1st plot
plt.plot(t, np.sin(t))

plt.subplot(1,2,2) # 1 row, 2 columns, 2nd plot
plt.plot(t, np.cos(t))

plt.tight_layout() # fixes the gap between plots
```

An overlay of 2-by-2 subplots

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

We can iterate over subplots to show several plots depending on a parameter:

```python
t = np.linspace(0, 2*np.pi, 100)
for i in range(1,7):
    plt.subplot(3,2,i) # three rows, 2 columns, i-th plot
    plt.plot(t, np.sin(i*t))
    plt.title(f"Plot of sin({i}t)")
plt.tight_layout()
    
```

# Histograms 

Generate normally distributed random numbers.

```python
x = np.random.randn(10000)
```

Display a histogram of the values in the array:

```python
plt.hist(x);
```

For more detail we can increase the number of bins.

```python
plt.hist(x, bins=50);
```

By default, `hist()` diplays frequency (count) values in each bin. If, instead, we want to see the fraction of values in each bin, we set the parameter `density=True`.

```python
plt.hist(x, bins=50, density=True);
```

This histogram should coincide with the bell curve from which the values are drawn.

Let us plot the normal distribution density $\frac{1}{\sqrt{2\pi}} \mathrm{e}^{-x^2/2}$ together with the histogram.

```python
plt.hist(x, bins=50, density=True, alpha=0.6) # increase the transparency of the histogram

t = np.linspace(-4,4,50)
y = np.exp(-t**2/2)/np.sqrt(2*np.pi)

plt.plot(t, y, lw=3);
```

# Bar graphs

Bar graphs show rectangular bars located at x-values, as high as the y-values.

```python
x = np.arange(20)
y = np.exp(-((x-10)/4)**2)  # a bell curve with mean 10 and spread 2
```

```python
plt.bar(x, y) # bar plot 
plt.plot(x, y, "ro-");  # line plot of the same data
```

Bar plots also accept values for error bars (need to be supplied separately)

```python
x = [1, 1.3, 1.7, 2, 2.2, 2.5, 3]
y = [0.4, 0.5, 0.7, 0.8, 0.8, 0.65, 0.45]
yerr = [0.05, 0.15, 0.1, 0.1, 0.08, 0.07, 0.09]

plt.bar(x, y, yerr=yerr, alpha=0.5)
```

To avoid overlap, bar widths can be set to smaller values.

```python
plt.bar(x,y, yerr=yerr, width=0.1)
```

# Scatter plots

Scatter plots display (x,y) pairs on the plot.

```python
# generate two arrays of random numbers
x = np.random.rand(30)
y = np.random.rand(30)
plt.scatter(x,y);
```

# Bubble plots

A *bubble plot* is a scatter plot where you use the size and the color of the points to show two more pieces of information.

With bubble plots, you can display four dimensions of information on two dimensional plots.

![](img/countries_health_wealth_2016_v13.jpg)

The `scatter()` function accepts parameters for the size and color of the points. We can generate bubble plots by using these parameters.

```python
size = np.random.randint(100,1001,30)
color = np.random.rand(30)
plt.scatter(x, y, s=size, c=color)
plt.colorbar();
```

The conversion from a numeric value to a color is done by means of a *color map*. Many color maps are available in matplotlib.

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.scatter(x, y, s=size, c=color, alpha=0.7)
plt.colorbar()

plt.subplot(1,3,2)
plt.scatter(x, y, s=size, c=color, cmap="jet", alpha=0.7)
plt.colorbar()

plt.subplot(1,3,3)
plt.scatter(x, y, s=size, c=color, cmap="gray", alpha=0.7)
plt.colorbar();
```
