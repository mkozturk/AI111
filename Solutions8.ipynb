## 1 - Table processing

A table of data is arranged as rows and columns, and each row corresponds to a data point. For example, consider the following table of grades of students from different subjects.

````python
s = """idno,name,math,literature,physics,music
27-6234266,Domenic Been,62,71,51,81
32-8500006,Carmel Vondrach,58,62,60,61
46-4848244,Gabey Stanlock,63,51,50,83
40-4452613,Miner Spoure,61,53,55,87
49-8047324,Joanie Padbery,49,60,51,68"""
````

The data is given as a single string. The table can be extended as we like.

### (a)
Given a string storing a table of grades, as shown above, write a program that prints the class average for every subject, and every student's average score.

Use basic string methods such as `.split()`, `.strip()` etc. to extract the data as relevant.

Use the first line to extract the column headers.

**Expected output:**
```
Class averages by subject
math 58.6
literature 59.4
physics 53.4
music 76.0

Student averages
27-6234266 Domenic Been 66.25
32-8500006 Carmel Vondrach 60.25
46-4848244 Gabey Stanlock 61.75
40-4452613 Miner Spoure 64.0
49-8047324 Joanie Padbery 57.0
```

````python
# solution
headers = s.split("\n")[0].split(",")
subjects = headers[2:]
data = s.split("\n")[1:]

# class averages in every subject
print("Class averages by subject")
for i,sub in enumerate(subjects):
    total = 0
    for row in data:
        lst = row.split(",")
        lst = lst[2:] # skip the id and name
        total += int(lst[i])
    print(sub, total/len(data))

print()
print("Student averages")
for row in data:
    lst = row.split(",")
    idno, name = lst[:2]
    grades = lst[2:]
    total = sum([int(x) for x in grades])
    print(idno, name, total/len(grades))
````

### (c)

Write a function `class_average(data, subject)` that returns the class average for a given subject (same as the column's header). The parameter `data` is a string storing a table of data, as shown in part (a).

Example:

````python
# solution
def class_average(data, subject):
    headers = data.split("\n")[0].split(",")
    table = data.split("\n")[1:]
    # get the column index of the subject
    subj_ind = headers.index(subject)
    total = 0
    for row in table:
        total += int(row.split(",")[subj_ind])
    return total/len(table)
````

````python
s = """idno,name,math,literature,physics,music
27-6234266,Domenic Been,62,71,51,81
32-8500006,Carmel Vondrach,58,62,60,61
46-4848244,Gabey Stanlock,63,51,50,83
40-4452613,Miner Spoure,61,53,55,87
49-8047324,Joanie Padbery,49,60,51,68"""

class_average(s, "math")
````

### (d)
Write a function `student_reports(data)` that returns a *nested dictionary* representing the same data, where keys are the id strings, and values are dictionaries with subjects and grades, as well as the student's average (see example below).

The function's parameter `data` is a string storing a table of data, as used above.

**Keep the function general.** It should not rely on the subject names given in the example. You can assume that the first columns are always "idno" and "name", but the other column names may be different.

Examples:

````python
# solution
def student_reports(s):
    headers = s.split("\n")[0].split(",")
    subjects = headers[2:]
    data = s.split("\n")[1:]
    result = dict()
    for row in data:
        row_lst = row.split(",")
        idno = row_lst[0]
        name = row_lst[1]
        grades = [int(x) for x in row_lst[2:]]
        result[idno] = {"name":name}
        result[idno].update(dict(zip(subjects,grades)))
        result[idno]["average"] = sum(grades)/len(grades)
    return result
````

````python
s = """idno,name,math,literature,physics,music
27-6234266,Domenic Been,62,71,51,81
32-8500006,Carmel Vondrach,58,62,60,61
46-4848244,Gabey Stanlock,63,51,50,83
40-4452613,Miner Spoure,61,53,55,87
49-8047324,Joanie Padbery,49,60,51,68"""

student_reports(s)
````

## 2 - Pick the numbers

The file `data_ex8_2.txt` contains a text with some numbers inserted between words. Download this file from the course GitHub site into your working directory.

Write a program that prints the **sum of all numbers** in the text.

**Expected output:**
```
89099
```

**Hints:**
* Read the entire file into a single string.
* Break the text into a list of words with the `split()` method.
* Check if each word consists solely of digits with the `isdigit()` method. If so, add the corresponding number to the total.
* Note that some numeric words will end with a period (`.`) or a comma (`,`). They should also be added up. Deal with them accordingly.

````python
# solution

with open("data_ex8_2.txt") as f:
    s = f.read()

total = 0
for x in s.split():
    if x.isdigit():
        total += int(x)
    elif x[:-1].isdigit():
        total += int(x[:-1])
print(total)
````

### (c) 
Write a function `total_revenue(channel, region, itemtype)` that returns the total revenue from customers from `channel` and `region` on the `itemtype` category of items.

**Examples:**

````python
# solution
def total_revenue(channel, region, itemtype):
    with open("data_ex8_3.csv") as f:
        header = f.readline().strip()
        item_index = header.split(",").index(itemtype)
        total = 0
        for line in f:
            row = line.strip().split(",")
            c = int(row[0])
            r = int(row[1])
            if c==channel and r==region :
                total += int(row[item_index])
    return total

print("Revenue Report")
print("==============")
for channel in [1,2]:
    print("Channel", channel)
    for region in [1,2,3]:
        print("\tRegion", region)
        for itemtype in ["Fresh","Milk","Grocery","Frozen","Detergents_Paper","Delicacies"]:
            print("\t\t",itemtype, total_revenue(channel, region, itemtype))
````

### (d)

Each row in the data indicates a unique customer. Write a function `best_customer(itemtype)` that returns the row where `itemtype` has the largest value.

Examples:

````python
# Solution
def best_customer(itemtype):
    with open("data_ex8_3.csv") as f:
        header = f.readline().strip()
        item_index = header.split(",").index(itemtype)
        maxval = 0
        
        for line in f:
            data = line.strip().split(",")
            if int(data[item_index]) > maxval:
                maxval = int(data[item_index])
                maxline = line
        return(maxline)
````
