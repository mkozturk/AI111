# Basic plotting with matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

# A simple plot

The most basic plot: Connect points with lines.

```python
x = [1,3,4,6,7,9]
y = [2,5,5,9,6,7]
plt.plot(x,y)
```

Plot a function

```python
x = np.linspace(0,10,200)
y = np.sin(x**1.5)
plt.plot(x,y)
```

# Size, labels, title, tick values, grid

Let's add axis labels.

```python
plt.xlabel("the independent variable")
plt.ylabel("the function value")
plt.title("The function sin(x^1.5)")
plt.plot(x,y)
```

Set the figure size explicitly.

```python
plt.figure(figsize=(12,5))
plt.xlabel("the independent variable")
plt.ylabel("the function value")
plt.title("The function sin(x^1.5)")
plt.plot(x,y)
```

```python
plt.figure(figsize=(12,5))
plt.xlabel("the independent variable")
plt.ylabel("the function value")
plt.title("The function sin(x^1.5)")
plt.grid(True)
plt.plot(x,y);
```

Modify the tick locations.

```python
plt.figure(figsize=(12,5))
plt.xlabel("the independent variable")
plt.ylabel("the function value")
plt.title("The function sin(x^1.5)")
plt.grid(True)

plt.xticks([0,1,2.1, 4.3, 6.5, 7.8, 9.5, 10])
plt.yticks(pl.linspace(-1,1,5))

plt.plot(x,y);
```

# Configuring the plot style

## Change the line style

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.plot(x, y, "--")
pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1),":")
pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```

Available line styles:
* dotted line `:`
* dashed line `--`
* dot-dash line `.-`
* straight line `-`

## Change line colors

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.plot(x, y, "b")  # blue solid line
pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1),"r--") # red dashed line
pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```

The following color abbreaviations ara available:

|character|color|
|---------|-----|
|`'b'`|	blue
|`'g'`|	green
|`'r'`	| red
|`'c'`|	cyan
|`'m'`|	magenta
|`'y'`|	yellow
|`'k'`|	black
|`'w'`|	white

Alternatively, we can specify colors by names (`'green'`), by hex strings (`'#008000'`), or by RGB or RGBA tuples ((0,1,0,1)). For more details, see [this page](https://matplotlib.org/users/colors.html).

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.plot(x, y, color="chocolate")

pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1), color=(0.5,0.1,0.5))

pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```

## Change line thickness

Use the `lw` or the `linewidth` keyword parameter. This specifies the thickness in units of points. Can be a float.

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.plot(x, y, lw=7)
pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1), linewidth=4)

pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```

## Setting the transparency

The `alpha` keyword argument sets the transparency level of a ploting object. 0.0 for completely transparent, 1.0 for completely opaque.

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.plot(x, y, lw=7, alpha=0.6)
pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1), linewidth=4, alpha=0.9)

pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```

## Configure the markers

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.plot(x, y, ".")  # Plot the data points without connecting them
pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1),"*-");  # Plot points with stars, connect with solid line

pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```

Markers can be combined in one format string specifying the color and line style.

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.plot(x, y, "rx")  # Plot the data points as red crosses without connecting them
pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1),"g+--");  # Plot points as green pluses, connect with dashed line

pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```

For a full list of available markers, see the documentation of `plot`

# Set axis limits

```python
pl.figure(figsize=(12,5))
pl.xlabel("the independent variable")
pl.ylabel("the function value")
pl.title("Two functions")
pl.grid(True)

pl.xlim(-2, 12) # x-axis values start at -2, end at 12
pl.ylim(-1.5, 3) # y-axis starts at -1.5, ends at 3

pl.plot(x, y)
pl.plot(x, pl.sin(5*x)/pl.sqrt(x+1));
pl.legend(["$\sin(x^{3/2})$", "$\sin(5x)/\sqrt{x+1}$"]);
```
